# Unveiling High-Ranking Website Attributes for Good Performance in Google SERP

## 1 - Introduction

### 1.1 - Executive Summary

The primary focus of this study is to understand how various website attributes and technologies influence a website's ranking on Google Search Engine Results Pages (SERP). Utilizing the comprehensive dataset from HTTP Archive, the research endeavors to pinpoint the metrics and characteristics integral to achieving superior ranking positions on Google. The aim is to uncover actionable insights for creating efficient, top-performing websites aligned with SEO, marketing, and business goals to enhance online visibility and economic efficiency in site development and maintenance.

**Key Findings:**

1. **Correlation Between Metrics and Website Rank:** The study reveals that no single metric strongly correlates with website rank; however, attributes like quick load times and a lower count of image requests have a negative correlation with rank, implying these factors could positively influence ranking.

2. **Impact of Certain Metrics on SEO:** Optimizations leading to improved website performance, such as faster load times and efficient data usage, are identified as positive contributors to better rankings. Efficiency in content delivery and adoption of modern web practices are highlighted as vital for competitive rankings.

3. **Technology's Influence on Rank:** The research suggests that the choice of technology has a notable impact on a website's rank, with higher-ranked sites utilizing technologies that boost performance, security, and informed decision-making.

4. **Page Load Performance:** A nuanced relationship is observed between the number of requests a webpage makes and its load performance. Fast-loading websites tend to have more requests, underscoring the importance of efficient request management over sheer quantity.

5. **Predictive Model Challenges:** Despite a promising model accuracy for rank prediction, challenges persist due to the dataset's imbalanced nature, especially for higher-ranked websites, and reliance on a singular dominant feature, indicating the complexity in reliably forecasting page ranks.

**Actionable Insights:**

- Emphasize optimizing site load times through reduced and efficiently handled requests, and strategic resource utilization.
  
- Efficient content and resource management strategies, including image compression, utilization of modern file formats, and optimization of code, are crucial.
  
- Adherence to modern web standards and practices and the integration of efficient web technologies are recommended to enhance user experience and SEO rankings.
  
- Incorporate data-driven insights into SEO strategies to prioritize optimization efforts aimed at improving user experience and consequently, rankings.

**Challenges and Future Directions:**

The study faces challenges due to an imbalanced dataset with underrepresented top-ranked websites and the exclusion of potentially influential data due to constraints. Future improvements involve balancing the dataset, expanding feature investigation, and incorporating additional insightful metrics, aiming for a more accurate and holistic understanding of website ranking factors.

This research serves as a foundational step towards optimizing website performance in alignment with Google's ranking preferences, offering a pathway for stakeholders to make informed decisions for enhanced online presence and operational efficiency.

### 1.4 - Data Source: HTTP Archive Dataset

**Data Source:** Dataset was downloaded from [httparchive.org](https://github.com/HTTPArchive/httparchive.org/tree/main)

The HTTP Archive is a project that keeps track of how websites are built https://httparchive.org/. It does this by regularly visiting popular websites and recording a bunch of data about them, including:

* The different resources that make up the website (like images, scripts, and stylesheets)
* How long it takes to load these resources
* The technologies that the website uses

This data is then stored and made available for anyone to download or analyze. Researchers, web developers, and anyone else interested in how websites work can use this data to learn more about the current state of the web and how it's changing over time.

#### 1.4.2 - How the Dataset relates to Google PageRank

While the dataset does not directly refer to a website's Google PageRank, there's a significant connection between the information it captures and the concepts underlying Google's approach to page experience and ranking. Google PageRank, historically, was a method to rank web pages in search engine results based on the quality and quantity of links to a page. However, Google's ranking algorithms have evolved to include a broader array of factors, notably focusing on the user experience (UX) on web pages. This shift aligns with what is provided in the HTTP Archive dataset.

Let's explore how the data captured by the HTTP Archive can relate to several aspects of page experience that Google considers important:

1. **Loading Performance (including SpeedIndex, TTFB, onLoad, fullyLoaded):** Google has made it clear that fast loading times are a crucial part of a good page experience. Metrics like Time to First Byte (TTFB), SpeedIndex, and the fullyLoaded time are key indicators of how quickly content becomes available to the user. These directly impact metrics such as Largest Contentful Paint (LCP), which is part of Google's Core Web Vitals.

2. **Interactivity and Visual Stability (onContentLoaded, visualComplete):** Interactivity, measured by First Input Delay (FID), and visual stability, captured by Cumulative Layout Shift (CLS), are critical Core Web Vitals. While the dataset does not explicitly mention FID or CLS, metrics like onContentLoaded and visualComplete can be proxies for understanding aspects of interactivity and visual stability, respectively.

3. **Mobile Friendliness (meta_viewport):** The presence and configuration of the viewport meta tag can influence how well a page scales on mobile devices, an aspect of mobile usability that Google factors into its rankings.

4. **Secure Browsing and HTTPS Usage (numHttps):** Security is a key component of user experience. Google rewards websites that use HTTPS—ensuring secure connections—which is captured by the number of resources fetched over HTTPS in this dataset.

5. **No Intrusive Interstitials:** While the dataset does not directly measure the presence of intrusive interstitials, features like `document_height` and `num_iframes` might indirectly hint at page layout practices that could affect this aspect of user experience.

Therefore, although the dataset from the HTTP Archive does not explicitly tie to Google PageRank, the breadth of web performance and page composition data it provides offers valuable insights into factors that Google considers important for page experience. These factors have a known impact on SEO and are instrumental in shaping the page ranking in search results. Analyzing based on these dataset metrics can align web page performance with the best practices recommended in Google's page experience guidelines.

### 5.1 - Performance metrics having the strongest correlation with website rank (Correlation Analysis)

#### 5.1.1 - Pearson Correlation

Pearson correlation is used to assess the linear relationship between continuous variables, making it a starting point for analyzing how different metrics may linearly relate to website ranks.

By using risk metrics defined in the section of Grouping Attributes, we calculate correlations between columns and rank for each metric.

````python
def calculate_correlation(dataframe, features):
    # Include 'rank' in the analysis by adding it to the list for the VectorAssembler
    all_features = ["rank"] + features
    
    # Assemble features into a vector column
    assembler = VectorAssembler(inputCols=all_features, outputCol="features")
    df_vector = assembler.transform(dataframe.select(*all_features))
    
    # Compute the correlation matrix
    matrix = Correlation.corr(df_vector, "features").head()
    
    # Extract the correlation matrix as a NumPy array
    correlation_matrix = matrix[0].toArray()
    
    # Extract only the correlations of features with 'rank', excluding the first element which is the correlation of 'rank' with itself
    correlation_with_rank = correlation_matrix[0, 1:]  # 0 is the index of 'rank', 1: skips the rank's self-correlation
    
    # Create a DataFrame to display correlations neatly
    correlation_df = pd.DataFrame(correlation_with_rank, index=features, columns=['Correlation with Rank'])
    return correlation_df
````

##### Normalization of 'page_load_metrics'

We proceed to normalize the group "page_load_metrics" to ensure that each feature contributes proportionally to the analysis, regardless of its scale. Otherwise, features with larger magnitudes may dominate the analysis.

> Used ChatGPT to help normalize the data

````python
metrics = ["TTFB", "renderStart", "onContentLoaded", "onLoad", "fullyLoaded", "visualComplete"]

# Clone the original df to a new DataFrame for modifications
df_normalized = df.select("rank", *metrics)

global_stats = df.select(
    [F.mean(column).alias(f"mean_{column}") for column in metrics] +
    [F.stddev(column).alias(f"std_{column}") for column in metrics]
).collect()[0]

# Apply normalization using the global mean and std for each metric on the new DataFrame
for metric in metrics:
    mean_val = global_stats[f"mean_{metric}"]
    std_val = global_stats[f"std_{metric}"]
    if std_val:  # To avoid division by zero
        df_normalized = df_normalized.withColumn(f"norm_{metric}", (F.col(metric) - mean_val) / std_val)
    else:
        df_normalized = df_normalized.withColumn(f"norm_{metric}", F.lit(0))  

# Group by rank in the normalized DataFrame and calculate the average of the normalized values
normalized_avg_by_rank = df_normalized.groupBy("rank").agg(
    *[F.avg(f"norm_{metric}").alias(f"avg_norm_{metric}") for metric in metrics]
).orderBy("rank")

# Collect the data for plotting
normalized_avg_data = normalized_avg_by_rank.collect()

# Extract data for plotting
ranks = [row['rank'] for row in normalized_avg_data]
normalized_metrics_values = [[row[f"avg_norm_{metric}"] for row in normalized_avg_data] for metric in metrics]

# Plotting
plt.figure(figsize=(12, 8))
bar_width = 0.15
x = np.arange(len(ranks))

for i, metric_values in enumerate(normalized_metrics_values):
    plt.bar(x + i * bar_width, metric_values, width=bar_width, label=metrics[i])

# Add labels, title, and legend
plt.xlabel("Rank")
plt.ylabel("Normalized Average Value")
plt.title("Normalized Average Performance Metrics by Rank")
plt.xticks(x + bar_width * (len(metrics) - 1) / 2, ranks)
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust layout to fit everything
plt.show()
````

The normalized analysis of these metrics showed distinct trends relative to website rank: for example, TTFB and renderStart exhibit an increasing trend as rank rises (worsens), suggesting longer delays in serving initial content and rendering for higher-ranked websites. In contrast, FullyLoaded and visualComplete show a decreasing trend, indicating faster loading and visual rendering times for higher-ranked sites. These trends underscore the correlation between website performance and rank, with faster-loading sites generally achieving better ranks, underscoring the role of performance in user satisfaction and search engine rankings.

To explore the relationship between the technologies used on websites and their page ranks, we perform a join operation between the `technologies` table and the `summary_pages` table based on the `url` field. #### 5.3.1 - Joining the Tables

|Column Name|Description|Type|

**SQL Query for the join**
```
CREATE OR REPLACE TABLE `is843-team4-project.httparchive.2024_03_01_desktop_technology` AS
SELECT s.url, s.rank, t.category, t.app
FROM `httparchive.summary_pages.2024_03_01_desktop` as s
INNER JOIN `httparchive.technologies.2024_03_01_desktop` as t
ON s.url = t.url
```

This SQL query creates a new table named `2024_03_01_desktop_technology` in the `is843-team4-project.httparchive` dataset.

After establishing a combined dataset of website ranks and the technologies used on these sites, we proceed to analyze the frequency of each technology category within different rank groups. This analysis helps identify the prevalence and distribution of specific technologies across ranked websites.

> Used ChatGPT for the cell below

````python
initial_setup = False
if initial_setup:
    spark = SparkSession.builder \
        .appName("RankTechnology") \
        .config("temporaryGcsBucket", "is843-team4-project") \
        .getOrCreate()

    file = 'is843-team4-project.httparchive.2024_03_01_desktop_technology'
    df_technology = spark.read.format('bigquery') \
        .option('table', file) \
        .load()

    # Calculate the total number of unique URLs for each rank
    rank_counts = df_technology.groupBy("rank").agg(countDistinct("url").alias("total_urls"))

    # Calculate the number of unique URLs for each combination of category, app, and rank
    category_app_rank_counts = df_technology.groupBy("category", "app", "rank").agg(countDistinct("url").alias("url_count"))

    # Calculate the proportion for each category, app, and rank combination
    proportion_data = category_app_rank_counts.join(rank_counts, "rank")\
        .withColumn("proportion", col("url_count") / col("total_urls"))

    # Prepare to count the distinct URLs per category, app, and rank
    url_counts = df_technology.groupBy("category", "app", "rank").agg(countDistinct("url").alias("count_urls"))

    # Combine the data for the pivot table
    combined_data = proportion_data.join(url_counts, ["category", "app", "rank"])

    # Use pivot to transform 'rank' into separate columns and calculate the max proportion and count for each rank
    result = combined_data.groupBy("category", "app").pivot("rank").agg(
        max_("proportion").alias("max_proportion"),
        max_("count_urls").alias("count_urls")  # Max of counts to get the actual count for each rank
    )

    result.show()
    result.write \
        .format("bigquery") \
        .option("table", "is843-team4-project.httparchive.2024_03_01_rank_technology") \
        .mode("overwrite") \
        .save()

file =  "is843-team4-project.httparchive.2024_03_01_rank_technology"
rank_df = spark.read.format('bigquery') \
    .option('table', file) \
    .load()
````

````python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, row_number, concat_ws


# Assuming Spark session is already started and rank_df is loaded
spark = SparkSession.builder.appName("Top N Ranks").getOrCreate()

# Define the rank columns and their corresponding labels
rank_columns = {
    "1000_count_urls": "rank1000",
    "5000_count_urls": "rank5000",
    "10000_count_urls": "rank10000",
    "50000_count_urls": "rank50000",
    "100000_count_urls": "rank100000",
    "500000_count_urls": "rank500000",
    "1000000_count_urls": "rank1000000",
    "5000000_count_urls": "rank5000000",
    "10000000_count_urls": "rank10000000",
    "50000000_count_urls": "rank50000000"
}

# Function to create a DataFrame with the top N for each rank using a window function
def top_n_for_all_ranks(n=20):
    final_df = None
    for column, label in rank_columns.items():
        window = Window.orderBy(col(column).desc())
        top_n_df = rank_df.withColumn("No", row_number().over(window)) \
                          .select(col("No"), 
                                  concat_ws(".", col("category"), col("app")).alias(label),
                                  col(column).alias(label + "_value")) \
                          .filter(col("No") <= n)
        if final_df is None:
            final_df = top_n_df
        else:
            final_df = final_df.join(top_n_df, "No", "outer")

    return final_df

# Compute the final DataFrame containing top N for all ranks
final_df = top_n_for_all_ranks()
# Display the final DataFrame
final_df.show(20, truncate=False)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
from pyspark.sql.functions import concat_ws, col

# Define a function to convert Spark DataFrame to a dictionary
def spark_df_to_dict(spark_df):
    return {col: spark_df.select(col).rdd.flatMap(lambda x: x).collect() for col in spark_df.columns}

# Convert Spark DataFrame to dictionary for plotting
data_dicts = [
    (rank_df.select(concat_ws(".", col("category"), col("app")).alias("category_app"), '1000_count_urls'), 'Top 10 Categories for rank 1000'),
    (rank_df.select(concat_ws(".", col("category"), col("app")).alias("category_app"), '100000_count_urls'), 'Top 10 Categories for rank 100000'),
    (rank_df.select(concat_ws(".", col("category"), col("app")).alias("category_app"), '50000000_count_urls'), 'Top 10 Categories for rank 50000000')
]

fig, axs = plt.subplots(1, 3, figsize=(20, 10))

for i, (df_, title) in enumerate(data_dicts):
    data_dict = spark_df_to_dict(df_)
    ax = axs[i]

    categories = data_dict['category_app']
    app_counts = data_dict[f'{title.split()[-1]}_count_urls']
    
    # Filter out None values from app_counts
    valid_data = [(cat, count) for cat, count in zip(categories, app_counts) if count is not None]
    
    # Sort by app_counts in descending order and take top 10
    valid_data.sort(key=lambda x: x[1], reverse=True)
    valid_data = valid_data[:10]
    
    # Unpack categories and app_counts
    valid_categories, valid_counts = zip(*valid_data)
    
    ax.bar(range(len(valid_categories)), valid_counts)
    ax.set_title(title)
    ax.set_xticks(range(len(valid_categories)))
    ax.set_xticklabels(valid_categories, rotation=90)
    ax.xaxis.set_major_locator(FixedLocator(range(len(valid_categories))))
    ax.set_xlabel('Category.App')
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
````

````python
train_df, test_df = df.randomSplit([0.7, 0.3], seed=843)
````

Define a list of features that we will exclude from the model fitting, with following reasons:
1. String value, which make it unfit for the model.
2. Unrelated column, such as url, date and id.

````python
non_feature_columns = ['pageid','crawlid','wptid', 'wptrun', 'urlhash', 'createDate', \
                       'startedDateTime', 'url', 'urlShort', 'archive', 'label', 'wptid', 'url', 'urlShort', 'cdn', 'doctype', 'meta_viewport', 'metadata']
````

````python
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.sql.types import StringType, TimestampType
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.sql.functions import col
````

````python
labelIndexer = StringIndexer(inputCol="rank", outputCol="indexedLabel").fit(train_df)

assembler = VectorAssembler(
    inputCols=[c for c in train_df.columns if c not in non_feature_columns], 
    outputCol="features")

rf = RandomForestClassifier(labelCol="indexedLabel", featuresCol="features", numTrees=10)

pipeline = Pipeline(stages=[labelIndexer, assembler, rf])

model = pipeline.fit(train_df)
````

````python
# #You can save the model for future use
# model.save("/path/to/save/model")
````

Predictions on the test df to test performance.

````python
predictions = model.transform(test_df)

evaluator = MulticlassClassificationEvaluator(
    labelCol="indexedLabel", predictionCol="prediction", metricName="accuracy")
````

#### 5.4.2 - Model performance

````python
accuracy = evaluator.evaluate(predictions)
precision = evaluator.setMetricName("weightedPrecision").evaluate(predictions)
recall = evaluator.setMetricName("weightedRecall").evaluate(predictions)
f1 = evaluator.setMetricName("f1").evaluate(predictions)

print("Test Accuracy = %g" % accuracy)
print("Weighted Precision = %g" % precision)
print("Weighted Recall = %g" % recall)
print("F1 Score = %g" % f1)
````

All measurements of model performance look promising, with accuracy as high as 0.93, but we could not conclude that this model does fit our dataset well for now.
